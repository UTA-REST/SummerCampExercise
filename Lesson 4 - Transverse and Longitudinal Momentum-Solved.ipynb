```python
#Imports etc
import matplotlib 
import numpy
import pylab
%matplotlib inline
```

In this session we will be looking to see if we can find hints of
any new heavy particles.

```python
# Load the full data sample
Events=numpy.loadtxt("Events.txt")
```

```python
# Calculate and histogram the longitudinal and transverse momenta 
#  of the first and second leptons in each event.
#  use 50 bins between 0 and 500 GeV
pylab.figure(figsize=(5,5))

Trans1=(Events[:,2]*Events[:,2]+Events[:,3]*Events[:,3])**0.5
Long1 = Events[:,4]
E1=Events[:,5]
Trans2=(Events[:,8]*Events[:,8]+Events[:,9]*Events[:,9])**0.5
Long2 = Events[:,10]
E2=Events[:,11]

pylab.hist(Trans1,bins=numpy.arange(0,500,10),label='Highest pt',histtype='step')
pylab.hist(Trans2,bins=numpy.arange(0,500,10),label='Second pt',histtype='step')
pylab.xlabel("pT/GeV",fontsize=14)
pylab.ylabel("number",fontsize=14)
pylab.legend(loc=1,fontsize=14)
pylab.show()

pylab.figure(figsize=(5,5))

pylab.hist(Long1,bins=numpy.arange(0,500,10),label='Highest pt',histtype='step')
pylab.hist(Long2,bins=numpy.arange(0,500,10),label='Second pt',histtype='step')
pylab.xlabel("pL/GeV",fontsize=14)
pylab.ylabel("number",fontsize=14)
pylab.legend(loc=1,fontsize=14)
pylab.show()
```

```python
# Plot the transverse momentum of lepton 1 vs the transverse momentum
#  of lepton 2.  Use scales 0 to 500 GeV on both axes
pylab.figure(figsize=(5,5))
pylab.plot(Trans1,Trans2,'.')
pylab.xlim(0,500)
pylab.ylim(0,500)
pylab.xlabel("pT of lepton 1",fontsize=14)
pylab.ylabel("pT of lepton 2",fontsize=14)
pylab.show()
```

```python
# Plot the higest pT for each of the subsamples
# you made in the last exercise.  Do you notice anything interesting?

EE=numpy.loadtxt("ee.txt")
EMu=numpy.loadtxt("emu.txt")
MuMu=numpy.loadtxt("mumu.txt")

pylab.figure(figsize=(5,5))

TransEE=(EE[:,2]*EE[:,2]+EE[:,3]*EE[:,3])**0.5
TransEMu=(EMu[:,2]*EMu[:,2]+EMu[:,3]*EMu[:,3])**0.5
TransMuMu=(MuMu[:,2]*MuMu[:,2]+MuMu[:,3]*MuMu[:,3])**0.5

pylab.hist(TransEE,bins=numpy.arange(0,500,10),label='e e',histtype='step')
pylab.hist(TransEMu,bins=numpy.arange(0,500,10),label='e mu',histtype='step')
pylab.hist(TransMuMu,bins=numpy.arange(0,500,10),label='mu mu',histtype='step')
pylab.xlabel("pT/GeV",fontsize=14)
pylab.ylabel("number",fontsize=14)
pylab.legend(loc=1,fontsize=14)
pylab.show()

pylab.figure(figsize=(5,5))
pylab.show()
```

```python
# Repeat for the (++ --), (+-, -+) charge subsets.  Anything interesting?

EESameSign=[]
EEOppSign=[]
EMuSameSign=[]
EMuOppSign=[]
MuMuSameSign=[]
MuMuOppSign=[]

for E in EE:
    if E[6]*E[12]==1:
        EESameSign.append(E)
    else:
        EEOppSign.append(E)
        
for E in EMu:
    if E[6]*E[12]==1:
        EMuSameSign.append(E)
    else:
        EMuOppSign.append(E)
    
for E in MuMu:
    if E[6]*E[12]==1:
        MuMuSameSign.append(E)
    else:
        MuMuOppSign.append(E)
    
EESameSign=numpy.array(EESameSign)
EEOppSign=numpy.array(EEOppSign)
EMuSameSign=numpy.array(EMuSameSign)
EMuOppSign=numpy.array(EMuOppSign)
MuMuOppSign=numpy.array(MuMuOppSign)
MuMuSameSign=numpy.array(MuMuSameSign)

pylab.figure(figsize=(5,5))

TransEESame=(EESameSign[:,2]*EESameSign[:,2]+EESameSign[:,3]*EESameSign[:,3])**0.5
TransEEOpp=(EEOppSign[:,2]*EEOppSign[:,2]+EEOppSign[:,3]*EEOppSign[:,3])**0.5
TransEMuSame=(EMuSameSign[:,2]*EMuSameSign[:,2]+EMuSameSign[:,3]*EMuSameSign[:,3])**0.5
TransEMuOpp=(EMuOppSign[:,2]*EMuOppSign[:,2]+EMuOppSign[:,3]*EMuOppSign[:,3])**0.5
TransMuMuSame=(MuMuSameSign[:,2]*MuMuSameSign[:,2]+MuMuSameSign[:,3]*MuMuSameSign[:,3])**0.5
TransMuMuOpp=(MuMuOppSign[:,2]*MuMuOppSign[:,2]+MuMuOppSign[:,3]*MuMuOppSign[:,3])**0.5

pylab.figure(figsize=(5,5))

pylab.hist(TransEESame,bins=numpy.arange(0,500,10),label='e e',histtype='step')
pylab.hist(TransEMuSame,bins=numpy.arange(0,500,10),label='e mu',histtype='step')
pylab.hist(TransMuMuSame,bins=numpy.arange(0,500,10),label='mu mu',histtype='step')

pylab.title("Same sign dileptons")
pylab.xlabel("pT/GeV",fontsize=14)
pylab.ylabel("number",fontsize=14)
pylab.legend(loc=1,fontsize=14)
pylab.show()


pylab.figure(figsize=(5,5))

pylab.title("Opposite sign dileptons")
pylab.hist(TransEEOpp,bins=numpy.arange(0,500,10),label='e e',histtype='step')
pylab.hist(TransEMuOpp,bins=numpy.arange(0,500,10),label='e mu',histtype='step')
pylab.hist(TransMuMuOpp,bins=numpy.arange(0,500,10),label='mu mu',histtype='step')

pylab.xlabel("pT/GeV",fontsize=14)
pylab.ylabel("number",fontsize=14)
pylab.legend(loc=1,fontsize=14)

pylab.show()
```
