```python
# A line that starts with a hash sign is a comment.  It does not get read by the interpreter.
%matplotlib inline
import numpy as np
import matplotlib as mpl
import pylab
import scipy as scip
```

```python
# Load the data from a file
Data=numpy.loadtxt("muons.txt")
```

```python
# Lets look at the shape of this array
numpy.shape(Data)
```

```python
# It has 11164 lines of 20 entries.  The 20 entries are 
# 0) event number
# 1) run number
# 2) particle 1 P_x
# 3) particle 1 P_y
# 4) particle 1 P_z
# 5) particle 1 E
# 6) particle 1 charge
# 7) particle 2 P_x
# 8) particle 2 P_y
# 9) particle 2 P_z
# 10) particle 2 E
# 11) particle 2 charge
# and then some more we won't be using in this exercise...

# print to screen the information for the first event

print (Data[0])
```

```python
# From this data, extract the p_t of particle 1 and particle 2 for each event.
# Note - you can do this either with a loop, or with numpy arrays.
# Label the axes, and include a legend

PTs1=np.sqrt(Data[:,2]*Data[:,2]+Data[:,3]*Data[:,3])
PTs2=np.sqrt(Data[:,7]*Data[:,7]+Data[:,8]*Data[:,8])
pylab.figure(figsize=(5,5))
pylab.hist(PTs1/1000,bins=100, histtype='step')
pylab.hist(PTs2/1000,bins=100, histtype='step')
pylab.xlabel("P_T / GeV")
pylab.ylabel("Number of events")
pylab.show()
```

```python
# Plot the PT of particle 1 against the PT of particle 2 in a scatter plot
pylab.figure(figsize=(5,5))
pylab.plot(PTs1/1000,PTs2/1000,'.')
pylab.xlabel("P_T1 / GeV")
pylab.ylabel("P_T2 / GeV")
pylab.show()
```

```python
InvMass=[]
px1s=[]
px2s=[]
py1s=[]
py2s=[]
pz1s=[]
pz2s=[]
E1s=[]
E2s=[]
m1s=[]
Count=0
for d in Data:
    Count=Count+1
    px1=d[2]
    py1=d[3]
    pz1=d[4]
    E1=d[5]
    px2=d[7]
    py2=d[8]
    pz2=d[9]
    E2=d[10]
    px1s.append(px1)
    px2s.append(px2)
    py1s.append(py1)
    py2s.append(py2)
    pz1s.append(pz1)
    pz2s.append(pz2)
    E1s.append(E1)
    E2s.append(E2)
    m1s.append(E1*E1-px1*px1-py1*py1-pz1*pz1)
    if(pow(E1+E2,2)-pow(px1+px2,2)-pow(py1+py2,2)-pow(pz1+pz2,2)<0):
        print "Bad!", Count
    else:
        InvMass.append(pow(pow(E1+E2,2)-pow(px1+px2,2)-pow(py1+py2,2)-pow(pz1+pz2,2),0.5))
```

```python
pylab.hist(m1s, bins=100,histtype='step')
pylab.semilogy()
```
